```python
year = [1972, 1982, 1992, 2002, 2012]
e_1 = [100.6, 158.61, 305.54, 394.3, 724.5]
e_2 = [10.5, 25.21, 58.65, 64.1, 100.3]
plt.plot(year, e_1, color = 'orange', label = 'e1')
plt.plot(year, e_2, color = 'g', label = 'e2', marker = 'D', markersize = 8, linewidth = 3, linestyle = '--') # '-' - solid, ':' - dotted, '-.' - dashed dot
plt.xlabel('Years')
plt.ylabel('eeeeeeeeee')
plt.legend()
plt.title('Energy')
plt.show()
```

```python
import numpy as np
fig = plt.figure()
ax = plt.axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
# fig.add_axes([0.1, 0.1, 0.8, 0.8]) # equivalent to above
x = np.linspace(-np.pi, np.pi, 15)
c = np.cos(x)
s = np.sin(x)
ax1 = ax.plot(x, c, 'bs:')
ax2 = ax.plot(x, s, 'rH--', markersize = 10)
ax.set_title('vnibwi')
ax.legend(labels = ('cosine', 'sin'), loc = 'upper left')
plt.show()
```

```python
x = [1, 2, 3]
y1 = [0, 1, 0]
y2 = [1, 0 ,1]
fig, ax = plt.subplots(2)
ax[0].plot(x, y1)
ax[1].plot(x, y2)
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y1')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y2')
ax[0].set_title('X vs Y1')
ax[1].set_title('X vs Y2')
plt.tight_layout()
plt.show()
```

```python
x = np.arange(0.0, 2.0, 0.01)
y = 1 + np.sin(2 * np.pi * x)
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3)
ax1.plot(x, y, color = 'magenta', linestyle = '-', linewidth = 2, label = '1')
ax2.plot(x, y, color = 'r', linestyle = '-.', linewidth = 2, label = '2')
ax3.plot(x, y, color = 'b', linestyle = '--', linewidth = 2, label = '3')
ax4.plot(x, y, color = 'black', linestyle = '-', linewidth = 2, label = '4')
ax5.plot(x, y, color = 'orange', linestyle = '-.', linewidth = 2, label = '5')
ax6.plot(x, y, color = 'green', linestyle = '--', linewidth = 2, label = '6')
ax1.set_xlabel('x')
ax1.set_ylabel('y1')
ax2.set_xlabel('x')
ax2.set_ylabel('y2')
ax3.set_xlabel('x')
ax3.set_ylabel('y3')
ax4.set_xlabel('x')
ax4.set_ylabel('y4')
ax5.set_xlabel('x')
ax5.set_ylabel('y5')
ax6.set_xlabel('x')
ax6.set_ylabel('y6')
plt.legend(loc = 'upper right')
plt.tight_layout()
plt.show()
```
